# 1. Django has the Model View Template Architecture
->Django has a Model View Template Architecture(MVT).

->The MVT(Model View Template Architecture) is a software design pattern 

-> It's a collection of three essential components Model,View and Template

->These thress layers are responsible for different things and can be used independently




## Model

->The Model helps to handle database 

-> it's a data access layer,which contains the requried fields and behaviour of the data you're storing

->There's hardly an application without a database

-> The Django frameworks offically support four databases ,namely PostgreSQL, MySQL, SQLite and oracle

-> A Model is a python class ,and it does not know anything about other Django layers,a application programming interface (API) is the only  means used to communicate between layers

->Models helps developers to  <B>Create,Read,Update,Delete objects (CRUD operations) </B> in the original database

-> Also ,they holds business logic , custom methods,properties and other things  related to the
data manipulation





## View 

-> View is used to execute the business logic and interact with the model to carry data and render a template 

->Views fetches data from the model(database).Then ,it either gives each template access to the specified data to be displayed ,or it processes data beforehand

->It accpet HTTP requests,applies business logic provided by Python classes and methods and provides HTTP responses to the client requests





## Template 

-> The template is a presentation layer that handles the User Interface part completely.

->These are files with HTML code ,which is used to render data

->The content of these files can be static or dynamic 

-> A template is used only to represent data since there's no business logic in it
